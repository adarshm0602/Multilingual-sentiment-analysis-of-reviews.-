# 04 · Classifier Benchmark

Compare **Naive Bayes**, **Logistic Regression**, and **Linear SVM** on Kannada-script product reviews using character n-gram TF-IDF features.

Run the benchmark script from the project root:

```bash
python scripts/build_eval_dataset.py
python scripts/run_classifier_benchmark.py
```

Results are saved to `data/processed/classifier_benchmark_results.json`.

````python
import json
import sys
from pathlib import Path

import pandas as pd

PROJECT_ROOT = Path.cwd().parent
sys.path.insert(0, str(PROJECT_ROOT))

from src.evaluation.classifier_benchmark import run_classifier_benchmark
from src.evaluation.dataset import load_indic_sentiment_kn, load_multilingual_eval_set

RESULTS_PATH = PROJECT_ROOT / 'data' / 'processed' / 'classifier_benchmark_results.json'

kn_df = load_indic_sentiment_kn()
print(f'Kannada benchmark set: {len(kn_df)} reviews')
kn_results = run_classifier_benchmark(kn_df)

rows = []
for name, metrics in kn_results['classifiers'].items():
    rows.append({
        'Classifier': name.replace('_', ' ').title(),
        'Accuracy': f"{metrics['accuracy']:.1%}",
        'Macro F1': f"{metrics['macro_f1']:.1%}",
        'Weighted F1': f"{metrics['weighted_f1']:.1%}",
    })

summary = pd.DataFrame(rows).sort_values('Macro F1', ascending=False)
summary
````

````python
print('Best classifier:', kn_results['best_classifier'])
print('Dataset size :', kn_results['dataset_size'])
print('Train / test :', kn_results['train_size'], '/', kn_results['test_size'])
print()
print(kn_results['classification_report'])

RESULTS_PATH.parent.mkdir(parents=True, exist_ok=True)
with RESULTS_PATH.open('w', encoding='utf-8') as fh:
    json.dump({'benchmarks': {'kannada_script_indic_sentiment': kn_results}}, fh, indent=2)
print('Saved', RESULTS_PATH)
````
